```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Online Retail.xlsx')
```

```python
df.head()
```

```python
# Quantos objetos do conjunto de dados possuem o valor do atributo InvoiceNo ausente ou nulo?
df['InvoiceNo'].isnull().value_counts()
```

```python
# Quantos objetos do conjunto de dados possuem o valor do atributo UnitPrice menor que zero? Apague esses objetos do conjunto de dados.
df[df['UnitPrice'] < 0]['UnitPrice'].count()
```

```python
df = df.drop(df[df['UnitPrice'] < 0].index)
df[df['UnitPrice'] < 0]['UnitPrice'].count()
```

```python
# Qual o cliente (atributo CustomerId) que comprou mais itens, sendo que cada linha representa um item comprado?
df['CustomerID'].value_counts().head(1)
```

```python
# Qual o produto mais caro vendido?
df = df.drop((df[df['Quantity'] < 0]).index)
df[df['UnitPrice'] == df['UnitPrice'].max()]
```

```python
#  Quais os pa√≠ses (Country) que fazem parte deste conjunto de dados?
df.groupby('Country').count()
```

```python
#  Qual a compra (InvoiceNo) que possui mais itens?
df['InvoiceNo'].value_counts().head(1)
```

```python
# Qual a compra (InvoiceNo) que tem o maior valor de compra?
df['Precototal'] = df['Quantity'] * df['UnitPrice']
df[df['Precototal'] == df['Precototal'].max()]
```
